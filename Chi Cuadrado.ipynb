# Gamma $(1, \frac 1 2) \equiv$  Exp $(\frac 1 2)$

Una exponencial de parámetro $\lambda = \frac 1 2$ tiene la misma distribución que una Gamma $(\alpha=1, \lambda= \frac 1 2)$

> Gamma$(\alpha, \lambda)$:
> $$\large f_X(x) = \frac {\lambda^\alpha}{\Gamma(\alpha)} x^{\alpha-1} e^{-\lambda x}$$

Si $\alpha = 1$:

* $\Gamma(\alpha) = 1$

> Gamma$(1, \lambda) \equiv$  Exp $(\lambda)$:
> $$\large f_X(x) = \lambda e^{-\lambda x}$$



```r
grilla <- seq(0, 15, 0.1)
# Gamma
plot(dgamma(grilla, 1, 1/2), pch=1, col="blue")

# Exponencial
points(dexp(grilla, 1/2), pch=1, cex=0.5, col="red")

# Chi Cuadrado
points(dchisq(grilla, df=2),cex=1.5, col="black")
```
