Densidad (*"longitud de ejes"*):

$$\large f(x;\theta) = \frac 1 \theta x ^ {(\frac 1 \theta - 1 )} \mathbb{1}_{[0,1]}(x)$$

```r
f <- function(x, theta){
    return ((1/theta) * x ^ (1/theta - 1))
}
```

# Visualizaciones de la Densidad

para distintos $\theta$

```r
gridDensidad <- function(theta, cant){
    longitudes <- numeric(cant)
    i <- 1
    for(x in seq(0, 1, length.out=cant)){
        longitudes[i] <- f(x, theta)
        i <- i+1
    }
    return (longitudes)
}
```

# Estimadores

Hecho en papel:

|Estimador de Máxima Verosimilitud|
|---|
|$$\huge \hat \theta_{MV} = - \frac {\sum_{i=1}^n X_i}{n}$$|

```r
n <- length(ejes)
theta_hat_MV <- - sum(log(ejes)) / n
theta_hat_MV
```

Hecho en papel:

|Estimador de Momentos|
|---|
|$$\huge \hat \theta_{m} = \frac {n}{\sum_{i=1}^n X_i} - 1$$|

```r
theta_hat_m <- n/sum(ejes) - 1
theta_hat_m
```

En plots:

```r
hist(ejes, prob=TRUE, col='steelblue')
abline(v=mean(ejes), col='red')
abline(v=theta_hat_m, col='blue')
abline(v=theta_hat_MV, col='orange')

plot(ejes, col='steelblue')
abline(h=mean(ejes), col='red')
abline(h=theta_hat_m, col='blue')
abline(h=theta_hat_MV, col='orange')
```
