**a ) i ) Media:**

Sumar una constante a **cada valor de la muestra** es como sumarle una constante **al promedio de la muestra**:

$$\bar X = \frac {\sum_{i=1}^n X_i} n $$

Si sumo constante $c$:

$$\bar Y = \frac {\sum_{i=1}^n X_i + c} n =$$

Saco $c$ afuera $n$ veces

$$= \frac {n . c + \sum_{i=1}^n X_i} n$$

$$= c + \frac {\sum_{i=1}^n X_i} n$$

$$= c + \bar X$$

Por lo que:

| Respuesta |
|---|
| $$\huge Y_i = X_i + c \implies \bar Y = \bar X + c$$ |

**Mediana:**

Con la mediana es igual, pero más simple:

* Si tengo una muestra de n elementos, 
  
  y sumo $c$ a cada uno de los elementos
  
  desplazo cada uno de los elementos en la magnitud y sentido de $c$,
  
  por lo que el elemento ***del medio*** de la muestra (ordenada) antes de sumarle $c$
  
  seguirá estando en su posición relativa con respecto a todos los otros elementos,
  
  pero con magnitud modificada por $c$
  
  
* Formalmente no se me ocurre, pero con números, sea la muestra ordenada:
  
  $$x_{(1)}, x_{(2)}, \dots, x_{(n/2)}, \dots, x_{(n-1)}, x_{(n)}$$
  
  Con $x_{(n/2)}$ la mediana de la muestra
  
  Si sumo $c$ a cada elemento, no cambio el órden relativo:
  
  $$x_{(1)}+c, x_{(2)}+c, \dots, x_{(n/2)}+c, \dots, x_{(n-1)}+c, x_{(n)}+c$$
  
  Con lo que se observa que la mediana es ahora: $x_{(n/2)}+c$
  
Por lo tanto:

| Respuesta |
|---|
| $$\huge Y_i = X_i + c \implies \tilde Y = \tilde X + c$$ |

**ii )** Con el producto de c es lo mismo, pero en vez de desplazar el rango, lo "estira/comprime"

La demo es igual.


| Respuesta |

| Respuesta |
|---|
| $$\huge Y_i = c X_i \implies \tilde Y = c \tilde X$$ |

**i )** 

$$S_X^2 = \frac 1 {n-1} \sum_{i=1}^n ( X_i - \bar X )^2$$

$$S_Y^2 = \frac 1 {n-1} \sum_{i=1}^n ( Y_i - \bar Y )^2$$

Si $Y_i = X_i + c$

$$S_Y^2 = \frac 1 {n-1} \sum_{i=1}^n ( X_i + c - (\bar X + c) )^2$$

$$= \frac 1 {n-1} \sum_{i=1}^n ( X_i - \bar X + c - c )^2$$

$$= \frac 1 {n-1} \sum_{i=1}^n ( X_i - \bar X)^2$$

>| Que es lo mismo que |
>|---|
>|$$\huge S_Y^2 = S_X^2$$|

**ii )**

$$S_Y^2 = \frac 1 {n-1} \sum_{i=1}^n ( Y_i - \bar Y )^2$$

Si $Y_i = c X_i$

$$S_Y^2 = \frac 1 {n-1} \sum_{i=1}^n ( c X_i - \bar c X_i )^2$$

$$= \frac 1 {n-1} \sum_{i=1}^n ( c (X_i - \bar X_i ))^2$$

$$= \frac 1 {n-1} \sum_{i=1}^n c^2 (X_i - \bar X_i )^2$$

$$= \frac 1 {n-1} c^2 \sum_{i=1}^n  (X_i - \bar X_i )^2$$

$$= c^2 \frac 1 {n-1}  \sum_{i=1}^n  (X_i - \bar X_i )^2$$

>| Que es lo mismo que |
>|---|
>|$$\huge  S_Y^2 = c^2 S_X^2$$|

**iii )**

$$S_X^2 = \frac 1 {n-1} \sum_{i=1}^n X_i^2 - \frac n {n-1} \bar X^2$$



Sé que 

$$S_X^2 = \frac 1 {n-1} \sum_{i=1}^n ( X_i - \bar X )^2$$

Abro el cuadrado

$$= \frac 1 {n-1} \sum_{i=1}^n ( X_i - 2 X_i \bar X^2 + \bar X^2 )$$

Separo la suma

$$= \left(  \frac 1 {n-1} \sum_{i=1}^n X_i^2 \right) - \left( \frac 1 {n-1} \sum_{i=1}^n 2 X_i \bar X \right) + \left( \frac 1 {n-1} \sum_{i=1}^n \bar X^2 \right) $$

Donde el promedio **no** depende de i:

> $$S_X^2 = \left(  \frac 1 {n-1} \sum_{i=1}^n X_i^2 \right) - \left( \frac 1 {n-1} \sum_{i=1}^n 2 X_i \bar X \right) + \left( \frac n {n-1} \bar X^2 \right) $$

Donde el primer término corresponde al primer término a probar.

Y el segundo:

$$\frac 1 {n-1} \sum_{i=1}^n 2 X_i \bar X =$$

$$= 2 \bar X \frac 1 {n-1} \sum_{i=1}^n X_i$$

Divido y multiplico por $n$ para que aparezca **otro** promedio $\bar X$

$$= 2 \bar X \frac 1 {n-1} n \frac {\sum_{i=1}^n X_i} n$$

$$= 2 \bar X \frac 1 {n-1} n \bar X $$

$$=  2 \frac {n} {n-1} \bar X^2$$

Reemplazando en la suma de antes:

> $$S_X^2 = \left(  \frac 1 {n-1} \sum_{i=1}^n X_i^2 \right) - \left( 2 \frac {n} {n-1} \bar X^2 \right) + \left( \frac n {n-1} \bar X^2 \right) $$

>| Con lo que obtenemos la relación a probar |
>|---|
>|$$\huge S_X^2 = \left(  \frac 1 {n-1} \sum_{i=1}^n X_i^2 \right) - \left( \frac {n} {n-1} \bar X^2 \right) $$|
