### Con datos generados en el sitio:

```r
data_n5 <- read.csv("./data/practica7/datos_Practica7_ej1_n_5.csv")
data_n50 <- read.csv("./data/practica7/datos_Practica7_ej1_n_50.csv")
data_n100 <- read.csv("./data/practica7/datos_Practica7_ej1_n_100.csv")
```

```r
# a) 
mean(data_n5$carbohidratos)
mean(data_n50$carbohidratos)
mean(data_n100$carbohidratos)
```

```r
# b)
median(data_n5$carbohidratos)
median(data_n50$carbohidratos)
median(data_n100$carbohidratos)
```

```r
#c) Proporcion de valores <= 76.5
length(data_n5$carbohidratos[data_n5$carbohidratos <= 76.5]) / length(data_n5$carbohidratos)
length(data_n50$carbohidratos[data_n50$carbohidratos <= 76.5]) / length(data_n50$carbohidratos)
length(data_n100$carbohidratos[data_n100$carbohidratos <= 76.5]) / length(data_n100$carbohidratos)
```
