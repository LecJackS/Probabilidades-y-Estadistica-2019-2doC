Let X be the weight in cenths of grams of female crickets of certain species.

We know that the mean of the population is $29 \times \frac {1}{100} grams$, and has a variance of $3\times 10^{-4}grams^2$ 

```r
mean <- 29
var  <- 3
```

Let's suppose X follows pretty closely a Chi squared distribution with 4 degrees of freedom moved 25 units to the right as:

$$X \sim \chi^2_4 + 25$$

```r
df = 4
```

```r
grid <- seq(0, 15, 0.1)
Chi <- dchisq(grid, df=df)
X_ <- Chi + 25
plot(grid+25, X_)
```

```r
X <- rchisq(n=100000, df=df) + 25
hist(X, freq=FALSE)
points(grid+25, Chi, pty=3)
abline(v=mean(X), col="steelblue", lty=2)
abline(v=df+ 25, col="orange", lty=3)
```

$$= P_\theta(X_1=x_1) P_\theta(X_2=x_2) \dots P_\theta(X_n=x_n)$$
