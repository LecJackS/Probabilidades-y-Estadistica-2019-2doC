Let $X$ be a Normal distributed random variable, with mean $\mu = 0$ and variance $\sigma^2 = 1$

$$X \sim \mathcal N(0,1)$$

```r
mean <- 0
var  <- 1
```

```r
grid <- seq(-5, 5, 0.1)
likelihood <- dnorm(grid, mean=0, sd=1)

plot(grid, likelihood,  
     xlab='Weight of cricket (in 100th of gram)',
     ylab='Likelihood')
```

```r
data <- rnorm(n=100000, mean=0, sd=1)
hist(data, freq=FALSE, col="steelblue")
points(grid, likelihood, pty=3)

abline(v=mean(data), col="steelblue", lty=2)
abline(v=1*sd(data)*c(-1,1), col="blue", lty=3)
abline(v=2*sd(data)*c(-1,1), col="orange", lty=3)
abline(v=3*sd(data)*c(-1,1), col="red", lty=3)
```

Let $X$ represent the weight in **cenths of grams** of female crickets of certain species.

We know that the mean of the population is $29 \times \frac {1}{100} grams$, and has a variance of $3\times 10^{-4}grams^2$ 

```r
mean <- 29
var  <- 3
```

Let's suppose X follows pretty closely a Chi squared distribution with 4 degrees of freedom moved 25 units to the right as:

$$X \sim \chi^2_4 + 25$$

```r
df = 4
```

```r
grid <- seq(0, 15, 0.1)
Chi <- dchisq(grid, df=df)
X <- Chi
plot(grid+25, X, 
     xlab='Weight of cricket (in 100th of gram)',
     ylab='Likelihood')
```

```r
data <- rchisq(n=100000, df=df) + 25
hist(data, freq=FALSE)
points(grid+25, Chi, pty=3)
abline(v=mean(data), col="steelblue", lty=2)
abline(v=df+ 25, col="orange", lty=3)
```

$$L(\theta | x_1, \dots, x_n) = P_\theta(X_1=x_1, X_2=x_2, \dots, X_n=x_n)$$

$X_i$ independent

$$= P_\theta(X_1=x_1) P_\theta(X_2=x_2) \dots P_\theta(X_n=x_n)$$

$$= \prod_{i=1}^n P_\theta(X_i=x_i)$$

$$= \prod_{i=1}^n \chi^2_\theta$$
